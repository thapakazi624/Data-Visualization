# Setup

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```


# Data

```python
# using insurance data available locally
insurance_data =pd.read_csv('insurance.csv')
```

```python
# seeing the data
insurance_data.head()
```

# Scatter Plot

```python
# scatter plotting column bmi against charges
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

```python
# scatter plotting column bmi against charges with regression line
sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

# Color Coded Scatter Plots

```python
# scatter plotting column bmi against charges differentiating with column smoker using color
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker'])
```

```python
# scatter plotting column bmi against charges, using lmplot for two regression lines for smoker column
sns.lmplot(x='bmi', y='charges', hue='smoker', data=insurance_data)
```

```python
# categorical scatter plotting column smoker against charges
sns.swarmplot(x=insurance_data['smoker'], y=insurance_data['charges'])
```
