# Setup

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

# Data

```python
# getting the museum visitors dataset locally
museum_data = pd.read_csv('Museum_Visitors.csv', index_col='Month', parse_dates=True)
```

```python
# basic look into the data
museum_data.head()
```

# plotting


```python
#plotting the data in a lineplot
plt.figure(figsize=(16,16))

sns.lineplot(data=museum_data)
plt.title('Monthly Visitors to Museums')
```

```python
# showing lineplot for particular columns
plt.figure(figsize=(16,16))

sns.lineplot(data=museum_data['America Tropical Interpretive Center'])
plt.title('America Tropical Interpretive Center')
```
