# Setup

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

# Data

```python
# iris data
iris_data = pd.read_csv('iris.csv', index_col='Id')
```

# Plots

```python
# line chart
plt.figure(figsize=(12,6))
sns.lineplot(data=iris_data)
```

```python
# styling
sns.set_style('dark')
# line chart
plt.figure(figsize=(12,6))
sns.lineplot(data=iris_data)
```
