# Setup

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

# Data

```python
flight_delay_data = pd.read_csv('flight_delays.csv', index_col='Month')
```

```python
flight_delay_data.head()
```

# Bar Chart

```python
# canvas size
plt.figure(figsize=(10,6))

# title
plt.title('Delay by Month')

# seaborn bar plot for NK by month
sns.barplot(x=flight_delay_data.index, y=flight_delay_data['NK'])

# vertical axis label
plt.ylabel('Arrival Delay')
```

# Heatmap

```python
plt.figure(figsize=(10,6))

# title
plt.title('Delay by Month')

# seaborn heatmap plot for NK by month
sns.heatmap(data=flight_delay_data, annot=True)

# vertical axis label
plt.ylabel('Arrival Delay')
```
